```python
import os

from inference import run_face_verification, visualize_results

from src.config import Config
```

```python
config = Config()

# Configuration
DEMO_DIR = "photos"
PERSON_NAME = "zhenya"

print("Starting face verification demo...")
print(f"Person: {PERSON_NAME}")
print(f"Demo directory: {DEMO_DIR}")

# Verify directory structure
required_dirs = [os.path.join(DEMO_DIR, PERSON_NAME), os.path.join(DEMO_DIR, "others")]

for dir_path in required_dirs:
    if not os.path.exists(dir_path):
        print(f"Creating directory: {dir_path}")
        os.makedirs(dir_path, exist_ok=True)
```

```python
# Run verification
try:
    anchor_path, results, expected_labels, metrics = run_face_verification(
        demo_dir=DEMO_DIR,
        person_name=PERSON_NAME,
    )

    # Visualize results
    visualize_results(
        anchor_path=anchor_path,
        results=results,
        expected_labels=expected_labels,
        metrics=metrics,
        person_name=PERSON_NAME,
    )

    print("\nVerification completed successfully!")

except FileNotFoundError as e:
    print(f"Error: {e}")
    print("Please add the missing images and try again.")
except Exception as e:
    print(f"Unexpected error: {e}")
```
